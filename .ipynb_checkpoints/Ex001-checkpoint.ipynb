```python
describe_ID = dados['ID'].value_counts()
describe_ID.value_counts()
```

```python
dupeMask = describe_ID == 2
#dupeMask[:313]
dupeMask
```

```python
describe_ID.index[0:313]
```

```python
dupeIDs = describe_ID.index[dupeMask]
display(dupeIDs)
len(dupeIDs)
```

```python
duplicados = dados.loc[dados['ID'].isin(dupeIDs[0:3]),:]
duplicados
```

#### Matriz booleana do tamanho do DataFrame original onde encontraremos os valores que possuem 'zeros'.

```python
dadosMaskZero = dados == 0
dadosMaskZero
```

```python
featureMaskZero = dadosMaskZero.iloc[:,1:].all(axis=1)
display(featureMaskZero)
sum(featureMaskZero)
```

```python
dadosLimpos_deZeros = dados.loc[~featureMaskZero, :]
dadosLimpos_deZeros
```

```python
dadosLimpos_deZeros['ID'].nunique()
```

### <center> Examinando valores das características e da resposta para desenvolver o modelo repditivo ideal </center>

```python
#Verificando o tipo de dados do dataset
dados = dadosLimpos_deZeros
dados.info()
```

### Examinar a coluna PAY_1, pois ela apresenta seu tipo de dado String, a qual deveria ser Integer

### Conforme a saída, 0 não é documentados no dicionário de dados. Vamos explorar um pouco mais

```python
dados['PAY_1'].value_counts()
```

```python
PercentagemPAY_1 = dados['PAY_1'].value_counts(normalize="True") * 100
PercentagemPAY_1
```

### Encontramos dados não documentados 0 e -2 e dados ausentes 'Not available'. Por enquanto vamos remover os dados ausentes aplicando as mascaras booleanas

```python
DadosValidosPAY_1 = dados['PAY_1'] != 'Not available'
DadosValidosPAY_1.value_counts()
```

```python
DadosValidosPAY_1
```

```python
dadosLimpos_deVazios = dados.loc[DadosValidosPAY_1, :]
dadosLimpos2 = dadosLimpos_deVazios
dadosLimpos2.info()
```

### <center>Convertendo tipos de dados</center>

```python
dadosLimpos2['PAY_1'] = dadosLimpos2['PAY_1'].astype('int64')
dadosLimpos2[['PAY_1', 'PAY_2']].info()
```

```python
dadosLimpos2.info()
```
