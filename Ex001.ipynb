#### Matriz booleana do tamanho do DataFrame original onde encontraremos os valores que possuem 'zeros'.

```python
dadosMaskZero = dados == 0
dadosMaskZero
```

```python
featureMaskZero = dadosMaskZero.iloc[:,1:].all(axis=1)
display(featureMaskZero)
sum(featureMaskZero)
```

```python
dadosLimpos_deZeros = dados.loc[~featureMaskZero, :]
dadosLimpos_deZeros
```

```python
dadosLimpos_deZeros['ID'].nunique()
```
